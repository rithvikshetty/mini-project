```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from __future__ import print_function,division
from sklearn import datasets, svm
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
from sklearn.preprocessing import LabelEncoder
```

# Working On Toy Dataset
## Using Iris dataset available in sklearn
Here we'll be using the toy dataset to improvise/develop the code. In the next section we'll work on the real dataset

```python
iris = datasets.load_iris()
x = iris.data[:,:2]
y = iris.target
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
def testing(model):
    pred = model.predict(x_test)
    right_ones = 0
    for i in range(len(y_test)):
        if pred[i] == y_test[i]:
            right_ones+=1
    acc = (right_ones/len(x_test))*100
    cnf = confusion_matrix(y_test,pred)
    print(f"Accuracy: {np.round(np.sum(np.diag(cnf))/np.sum(cnf), 4) * 100 }%")
    print(cnf)
    return acc
```

```python
kernels = ('linear','poly','rbf')
accuracy_arr = []
i=0
for kernel in kernels:
    model = svm.SVC(kernel=kernel)
    model.fit(x_train,y_train)
    acc = testing(model)
    accuracy_arr.append(acc)
    print(f"Kernel : {kernel}\t,  Accuracy by hand : {acc}%")
    print("========================================")
    
```

# Working with Actual Data

```python
data = pd.read_csv("Iris.csv")
data = data.drop("Id",axis=1)
print(data.shape)
```

```python
data.describe()
```

#### Checking for Null data

```python
data.isnull().sum()
```

#### Traning and Testing Data splitting

```python
x = data[["SepalLengthCm","SepalWidthCm"]]
y = data["Species"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

#### Execution

```python
kernels = ['linear','poly','rbf']
for kernel in kernels :
    model = svm.SVC(kernel=kernel)
    model.fit(x_train,y_train)
    predictions = model.predict(x_test)
    #acc = accuracy_score(y_test,predictions)
    cnf = confusion_matrix(y_test,predictions)
    print(f"Kernel : {kernel} ,  Accuracy: {np.round(np.sum(np.diag(cnf))/np.sum(cnf), 4) * 100 }%\n",cnf)
```

## Output Visualizations

### Considering Sepal length and Sepal width

```python
x = data[["SepalLengthCm","SepalWidthCm"]]
y = data["Species"]

#convert from pandas to numphy
x=x.to_numpy()
y=y.to_numpy()

#converting categorical values of y to numeric values
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)


#using linear
svm = SVC(C=0.5, kernel='linear')
svm.fit(x, y)

# Plotting decision regions
plot_decision_regions(x, y, clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('sepal width [cm]')
plt.title('SVM on Iris using linear kernel')
plt.show()



#using poly

svm = SVC(C=0.5, kernel='poly')
svm.fit(x, y)

# Plotting decision regions
plot_decision_regions(x, y, clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('sepal width [cm]')
plt.title('SVM on Iris using poly kernel')
plt.show()


#using rbf

svm = SVC(C=0.5, kernel='rbf')
svm.fit(x, y)

# Plotting decision regions
plot_decision_regions(x, y, clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('sepal width [cm]')
plt.title('SVM on Iris using rbf kernel')
plt.show()
```
