#### <b>What is SVM?</b>
##### <li> Support vector machine is another simple algorithm that every machine learning expert should have in his/her arsenal. </li>
##### <li> Support vector machine is highly preferred by many as it produces significant accuracy with less computation power. </li>
##### <li> Support Vector Machine, abbreviated as SVM can be used for both regression and classification tasks. But, it is widely used in classification objectives. </li>
##### <li> The main objective of the support vector machine algorithm is to find a hyperplane in an N-dimensional space(N — the number of features) that distinctly classifies the data points.</li>

#### <b>Methodology</b>
##### <li>To separate the two classes of data points, there are many possible hyperplanes that could be chosen. Our objective is to find a plane that has the maximum margin, i.e the maximum distance between data points of both classes. Maximizing the margin distance provides some reinforcement so that future data points can be classified with more confidence.</li>
##### <li>Support vectors are data points that are closer to the hyperplane and influence the position and orientation of the hyperplane. Using these support vectors, we maximize the margin of the classifier. Deleting the support vectors will change the position of the hyperplane. These are the points that help us build our SVM.</li>

This image shows the margins of a hyperplane. Usually, the higher the margin higher is the performance, i.e, accuracy:

#### <b>Kernels</b>
##### The Kernels are just different ways for calculating the dot product of two vectors `x` and `y` for multi-dimensional vector space. The way that we compute the dot product also matters a lot, since we have to take the projection on the support vectors for calculation of inter-margin distance.

## Necessary Imports for the requirements

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mlxtend
from __future__ import print_function,division
from sklearn import datasets, svm
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
from sklearn.preprocessing import LabelEncoder
```

# Working On Trial Dataset
## Using Iris dataset available in sklearn
Here we'll be using the toy dataset to improvise/develop the code. In the next section we'll work on the real dataset. The sole purpose of this dataset is to evaluate the performance of the code and to try and improve the accuracy as much as possible.

```python
iris = datasets.load_iris()
x = iris.data[:,:2]
y = iris.target
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
def testing(model):
    pred = model.predict(x_test)
    right_ones = 0
    for i in range(len(y_test)):
        if pred[i] == y_test[i]:
            right_ones+=1
    acc = (right_ones/len(x_test))*100
    cnf = confusion_matrix(y_test,pred)
    print(f"Accuracy: {np.round(np.sum(np.diag(cnf))/np.sum(cnf), 4) * 100 }%")
    print(cnf)
    return acc
```

```python
kernels = ('linear','poly','rbf')
accuracy_arr = []
i=0
for kernel in kernels:
    model = svm.SVC(kernel=kernel)
    model.fit(x_train,y_train)
    acc = testing(model)
    accuracy_arr.append(acc)
    print(f"Kernel : {kernel}\t,  Accuracy by hand : {acc}%")
    print("========================================")
    
```

# Working with CSV Data
Here we'll be making use of CSV based dataset and not from the `dataset` library of scikit

```python
data = pd.read_csv("Iris.csv")
data = data.drop("Id",axis=1)
print(data.shape)
```

```python
data.describe()
```

Since No `Null` values found we directly head to modelling.


```python
data.isnull().sum()
```

### Sepal Length and Sepal Width used for predicting the Specicies
Here we are using the attributes of `Sepal Length` and `Sepal Width` for classifying the data into 3 classes of species

#### Traning and Testing Data splitting

```python
x = data[["SepalLengthCm","SepalWidthCm"]]
y = data["Species"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

#### Execution

```python
kernels = ['linear','poly','rbf']
for kernel in kernels :
    model = svm.SVC(kernel=kernel)
    model.fit(x_train,y_train)
    predictions = model.predict(x_test)
    #acc = accuracy_score(y_test,predictions)
    cnf = confusion_matrix(y_test,predictions)
    print(f"Kernel : {kernel} ,  Accuracy: {np.round(np.sum(np.diag(cnf))/np.sum(cnf), 4) * 100 }%\n",cnf)
```

### Petal Length and Petal Width used for predicting the Specicies
Here we are using the attributes of `Petal Length` and `Petal Width` for classifying the data into 3 classes of species

```python
x = data[["PetalLengthCm","PetalWidthCm"]]
y = data["Species"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
kernels = ['linear','poly','rbf']
for kernel in kernels :
    model = svm.SVC(kernel=kernel)
    model.fit(x_train,y_train)
    predictions = model.predict(x_test)
    #acc = accuracy_score(y_test,predictions)
    cnf = confusion_matrix(y_test,predictions)
    print(f"Kernel : {kernel} ,  Accuracy: {np.round(np.sum(np.diag(cnf))/np.sum(cnf), 4) * 100 }%\n",cnf)
```

### Considering Sepal length and Sepal width

```python
x = data[["SepalLengthCm","SepalWidthCm"]]
y = data["Species"]

#convert from pandas to numphy
x=x.to_numpy()
y=y.to_numpy()

#converting categorical values of y to numeric values
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)


#using linear
svm = SVC(C=0.5, kernel='linear')
svm.fit(x, y)

# Plotting decision regions
plot_decision_regions(x, y, clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('sepal width [cm]')
plt.title('SVM on Iris using linear kernel')
plt.show()



#using poly

svm = SVC(C=0.5, kernel='poly')
svm.fit(x, y)

# Plotting decision regions
plot_decision_regions(x, y, clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('sepal width [cm]')
plt.title('SVM on Iris using poly kernel')
plt.show()


#using rbf

svm = SVC(C=0.5, kernel='rbf')
svm.fit(x, y)

# Plotting decision regions
plot_decision_regions(x, y, clf=svm, legend=2)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('sepal width [cm]')
plt.title('SVM on Iris using rbf kernel')
plt.show()
```

## Conclusion

#### All of the Machine Learning Models have their own limitations, so does SVM. I'd like to summarize this presentation with the merits and demerits of SVM model. Since our model has 3 classes SVM isn't exactly the best fit. But, still it has pretty good performance. We have used 2 sets of parameters for classification i.e, `Sepal Length` and `Sepal Width` also `Petal Length` and `Petal Width` and since the first dataset is scattered more than the second we get better performance results with the `Petal` data.
##### <i>Advantages :</i>
<li>It is really effective in the higher dimension.</li>
<li>Effective when the number of features are more than training examples.</li>
<li>Best algorithm when classes are separable.</li>
<li>SVM is suited for extreme case binary classification.</li>

##### <i>Disadvantages :</i>
<li>For larger dataset, it requires a large amount of time to process.</li>
<li>Does not perform well in case of overlapped classes.</li>
<li>Selecting, appropriately hyperparameters of the SVM that will allow for sufficient generalization performance.</li>
<li>Selecting the appropriate kernel function can be tricky.</li>
